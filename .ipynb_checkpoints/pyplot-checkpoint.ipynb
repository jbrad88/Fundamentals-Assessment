### Plot 1 - Line plot

To plot a line plot in Matplotlib, we use the "plot()" function from PyPlot. [https://stackabuse.com/matplotlib-line-plot-tutorial-and-examples/]


````python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0.0, 10.0, 0.01)
y = 3.0 * x + 1.0

plt.plot(x, y, 'b-')

# Add a title
plt.title("Simple Line Plot")
# Add X and Y labels
plt.xlabel("Weight")
plt.ylabel("Mass")

plt.show()
````

### Plot 2 - Scatter Plot
The scatter plot is similar to the line plot with one difference - instead of points being joined by line segments, the points are represented individually  [https://jakevdp.github.io/PythonDataScienceHandbook/04.02-simple-scatter-plots.html]


````python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0.0, 10.0, 0.01)
y = 3.0 * x + 1.0

noise = np.random.normal(0.0, 1.0, len(x))

plt.plot(x, y + noise, 'r.')

# Add a title
plt.title("Simple Scatter plot")
# Add X and Y labels
plt.xlabel("Weight")
plt.ylabel("Mass")

plt.show()
````

### Plot 3 - Histogram

A histogram is a graph which shows frequency distribution (i.e. the frequency of various outcomes in a sample) [https://www.w3schools.com/python/matplotlib_histograms.asp]. 

In Matplotlib, we use the "hist()" function to create a histogram. 


````python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(0.0, 1.0, 1000)
# Use Numpy to randomly generate an array with 1000 values,
# concentrated around 0 with a standard deviation of 1. 

plt.hist(x)

plt.show()
````
